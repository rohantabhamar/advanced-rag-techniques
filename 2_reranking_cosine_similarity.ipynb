```python
from langchain_chroma import Chroma
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
len(text_spliter)
```

```python
embedding_model = get_embeddings()
model = get_model()
```

```python
vectorstore=Chroma.from_documents(text_spliter,embedding_model)
vectorstore_retreiver = vectorstore.as_retriever(search_kwargs={"k": 3})
collection = vectorstore._collection
result = collection.get(include=["embeddings", "documents", "metadatas"])
```

```python
query = "BSBI"
query_vector = embedding_model.embed_query(query)

print(query_vector)  
```

```python
len(result['embeddings'])
```

```python
similarities = cosine_similarity([query_vector], result['embeddings'])[0]
```

```python
len(similarities)
```

```python

most_similar_index = np.argmax(similarities)
```

```python
most_similar_document = result['documents'][most_similar_index]
```

```python
most_similar_document 
```

```python
top_k = 4
top_indices = np.argsort(similarities)[::-1][:top_k]
contents = []

# Step 5: print results
for i, idx in enumerate(top_indices):
    print(f"Rank {i+1}")
    print(f"  Score : {similarities[idx]:.4f}")
    contents.append(result['documents'][idx])
    print(f"  Chunk : {(result['documents'][idx])}")
    print()
```
